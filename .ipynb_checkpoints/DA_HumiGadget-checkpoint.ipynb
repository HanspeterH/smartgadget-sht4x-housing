# Influence of the Housing Humi Gadget on Accuracy and Response Time

Data Analysis of the influence of the new Humi Gadget Housing on the accuracy and response time of changing temperatures and humidities using the thunder to generate a stable environment.

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
#from quickplot import QuickPlot

import sensirion_fastedf as fastedf
import datetime
import data_access as da
import csv
import math
from scipy.optimize import curve_fit

import os
print(os.getcwd())
```

```python
humi_gadget_full = fastedf.read_edf('/home/jovyan/humi-gadget-housing-data-analysis/Thunder_1/Sensirion_MyAmbience_SHT40_Gadget_4.edf')
SHT_85_1_full = fastedf.read_edf('/home/jovyan/humi-gadget-housing-data-analysis/Thunder_1/2022-09-22_09-38-24-SHT4x_256968205.edf')
SHT_85_2_full = fastedf.read_edf('/home/jovyan/humi-gadget-housing-data-analysis/Thunder_1/2022-09-22_09-38-24-SHT4x_256963899.edf')
DM_full = fastedf.read_edf('/home/jovyan/humi-gadget-housing-data-analysis/Thunder_1/20220922T193213Z_DPM373_DewPointMirror.edf')

start = datetime.datetime(2022,9,23,2,0).timestamp()

humi_gadget_full.Epoch_UTC = humi_gadget_full.Epoch_UTC - start
SHT_85_1_full.Epoch_UTC = SHT_85_1_full.Epoch_UTC - start
SHT_85_2_full.Epoch_UTC = SHT_85_2_full.Epoch_UTC - start
DM_full.Epoch_UTC = DM_full.Epoch_UTC - start

humi_gadget = humi_gadget_full.where(humi_gadget_full.Epoch_UTC > 0).dropna()
SHT_85_1 = SHT_85_1_full.where(SHT_85_1_full.Epoch_UTC > 0).dropna()
SHT_85_2 = SHT_85_2_full.where(SHT_85_2_full.Epoch_UTC > 0).dropna()
DM = DM_full.where(DM_full.Epoch_UTC > 0).dropna()
print(humi_gadget)
# print(test)
# test.head()
#print(humi_gadget.T)
#print(SHT_85_1)
#print(SHT_85_2)
#print(DM_full)
DM_full.RH_DewPointMirror.plot()
DM_full.T_DewPointMirror.plot()
```

```python
fig, axes = plt.subplots(2,1, figsize=(10,10))

ax = axes[0]
ax.set_title('Temperature Humi Gadget in Housing')
ax.plot(humi_gadget.Epoch_UTC/60, humi_gadget['T'], color='tab:green', label='Humi Gadget in Housing')
ax.plot(SHT_85_1.Epoch_UTC/60, SHT_85_1.T_SHT4x_256968205, color='gold', label='SHT 85')
ax.plot(SHT_85_2.Epoch_UTC/60, SHT_85_2.T_SHT4x_256963899, color='blue', label='SHT 85')
ax.plot(DM.Epoch_UTC/60, DM.T_DewPointMirror, color='red', label='Dew Point Mirror')
ax.set_ylabel('Temperature [°C]', fontsize=14)
# ax.set_ylim(24,26)

ax = axes[1]
ax.set_title('Humidity Humi Gadget in Housing')
ax.plot(humi_gadget.Epoch_UTC/60, humi_gadget.RH, color='tab:green', label='Humi Gadget in Housing')
ax.plot(SHT_85_1.Epoch_UTC/60, SHT_85_1.RH_SHT4x_256968205, color='gold', label='SHT 85')
ax.plot(SHT_85_2.Epoch_UTC/60, SHT_85_2.RH_SHT4x_256963899, color='blue', label='SHT 85')
ax.plot(DM.Epoch_UTC/60, DM.RH_DewPointMirror, color='red', label='Dew Point Mirror')
ax.set_ylabel('Relative Humidity [%]', fontsize=14)
ax.set_xlabel('Time [min]', fontsize=14)
# ax.set_ylim(50,55.5)

for ax in axes:
    ax.set_xlim(0,250)
    ax.legend(loc='upper right')
    ax.tick_params(labelsize=12)
    ax.grid()
```
